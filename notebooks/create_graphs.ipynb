```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/results/rnn_languages_results.csv')
```

```python
df.head()
```

```python
refined_df = df[['Edited GRU languages 1024 - batch_loss', 
                 'Original GRU languages 1024 - batch_loss', 
                 'Edited GRU languages 512 - batch_loss', 
                 'Original GRU languages 512 - batch_loss', 
                 'Edited GRU languages 256 - batch_loss', 
                 'Original GRU languages 256 - batch_loss']]
```

```python
refined_df.head()
```

```python
refined_df = refined_df.rename(columns={
    "Edited GRU languages 1024 - batch_loss":"Augmented - 1024",
    "Original GRU languages 1024 - batch_loss":"Default - 1024",
    "Edited GRU languages 512 - batch_loss":"Augmented - 512",
    "Original GRU languages 512 - batch_loss":"Default - 512",
    "Edited GRU languages 256 - batch_loss":"Augmented - 256",
    "Original GRU languages 256 - batch_loss":"Default - 256",
})
```

```python
df_1024 = refined_df[['Augmented - 1024', 
                        'Default - 1024']]
df_512 = refined_df[['Augmented - 512', 
                        'Default - 512']]
df_256 = refined_df[['Augmented - 256', 
                        'Default - 256']]
```

```python
x_len = refined_df.shape[0]
```

```python
x_len
```

```python
refined_df.plot()
```

```python
fig, axs = plt.subplots(figsize=(8, 4))        # Create an empty Matplotlib Figure and Axes
refined_df.plot.line(ax=axs, style={'Augmented - 1024':'+:r',
                                    'Default - 1024':'r',
                                    'Augmented - 512':'+:b',
                                    'Default - 512':'b',
                                    'Augmented - 256':'+:g',
                                    'Default - 256':'g',})                   # Use pandas to put the area plot on the prepared Figure/Axes
axs.set_ylabel("Batch Loss")
axs.set_xlabel("Epochs")
axs.set_title("Batch loss for RNNs on 'Languages' dataset")
fig.savefig("../data/results/rnn_languages_plot.png")           # Save the Figure/Axes using the existing Matplotlib method.
plt.show()      
```

Individual Plots

```python
fig, axs = plt.subplots(figsize=(8, 4))        # Create an empty Matplotlib Figure and Axes
df_256.plot.line(ax=axs, style={'Augmented - 256':'r',
                                    'Default - 256':'b',}, )                   # Use pandas to put the area plot on the prepared Figure/Axes
axs.set_ylabel("Batch Loss")
axs.set_xlabel("Epochs")
axs.set_title("Loss of RNN's with hidden size=256")
axs.legend(["Augmented", "Default"])
#fig.savefig("no2_concentrations.png")           # Save the Figure/Axes using the existing Matplotlib method.
plt.show()   
```
