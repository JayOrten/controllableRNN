```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import matplotlib.patches as mpatches
```

```python
df = pd.read_csv('../data/results/rnn_languages_results.csv')
```

```python
df.head()
```

```python
refined_df = df[['Edited GRU languages 1024 - batch_loss', 
                 'Original GRU languages 1024 - batch_loss', 
                 'Edited GRU languages 512 - batch_loss', 
                 'Original GRU languages 512 - batch_loss', 
                 'Edited GRU languages 256 - batch_loss', 
                 'Original GRU languages 256 - batch_loss']]
```

```python
refined_df.head()
```

```python
refined_df = refined_df.rename(columns={
    "Original GRU languages 1024 - batch_loss":"Default; 1024 hidden size",
    "Edited GRU languages 1024 - batch_loss":"Alternative; 1024 hidden size",
    "Original GRU languages 512 - batch_loss":"Default; 512 hidden size",
    "Edited GRU languages 512 - batch_loss":"Alternative; 512 hidden size",
    "Original GRU languages 256 - batch_loss":"Default; 256 hidden size",
    "Edited GRU languages 256 - batch_loss":"Alternative; 256 hidden size",
})
```

```python
refined_df.shape
```

```python
refined_df = refined_df[refined_df.index % 8 == 0]
```

```python
refined_df
```

```python
df_1024 = refined_df[['Augmented - 1024', 
                        'Default - 1024']]
df_512 = refined_df[['Augmented - 512', 
                        'Default - 512']]
df_256 = refined_df[['Augmented - 256', 
                        'Default - 256']]
```

```python
x_len = refined_df.shape[0]
```

```python
x_len
```

```python
refined_df.plot()
```

```python
fig, axs = plt.subplots(figsize=(8, 4))        # Create an empty Matplotlib Figure and Axes
refined_df.plot.line(ax=axs, style={'Default; 1024 hidden size':'C0',
                                    'Alternative; 1024 hidden size':'+:C0',
                                    'Default; 512 hidden size':'C1',
                                    'Alternative; 512 hidden size':'+:C1',
                                    'Default; 256 hidden size':'C2',
                                    'Alternative; 256 hidden size':'+:C2',})                   # Use pandas to put the area plot on the prepared Figure/Axes
axs.set_ylabel("Batch Loss")
axs.set_xlabel("Epochs")
axs.set_title("Batch loss for RNNs on 'Languages' dataset")
axs.set_xlim(0, 145)
axs.set_ylim(0, 11)
# Setup custom legend
# Guide: https://stackoverflow.com/questions/39500265/how-to-manually-create-a-legend
line = Line2D([0], [0], label='Default', linestyle='solid', color='k')
line_2 = Line2D([0], [0], label='Alternative', linestyle='dashed', color='k')
books_2_patch = mpatches.Patch(color='C0', label = '1024 hidden size')
books_3_patch = mpatches.Patch(color='C1', label = '512 hidden size')
books_6_patch = mpatches.Patch(color='C2', label = '256 hidden size')
handles, labels = fig.gca().get_legend_handles_labels()
#handles.extend([books_2_patch])
plt.legend(handles=[line, line_2, books_2_patch, books_3_patch, books_6_patch])
fig.savefig("../data/results/rnn_languages_plot.png",bbox_inches='tight')           # Save the Figure/Axes using the existing Matplotlib method.
plt.show()      
```
