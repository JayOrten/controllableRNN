Breakthrough with books_6_1024 2 and 8, loss

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import matplotlib.patches as mpatches
```

```python
df_books_6 = pd.read_csv('../../data/results/july_7_results/books_6_loss_breakthrough.csv')
```

```python
df_books_6.head()
```

```python
df_books_6 = df_books_6.drop('Step', axis=1)
```

```python
df_books_6 = df_books_6.iloc[:, [int(x) for x in range(0, 12, 3)]]
```

```python
df_books_6.head()
```

```python
fig, axs = plt.subplots(figsize=(5,4))
df_books_6.plot.line(ax=axs, style={'edited_3_books_6_1024_8 - train_loss':':C0',
                                'edited_3_books_6_1024_2 - train_loss':':C1',
                                'normal_books_6_1024_8 - train_loss':'C0',
                                'normal_books_6_1024_2 - train_loss':'C1',})
axs.set_ylabel("Batch Loss")
axs.set_xlabel("Epochs")
axs.set_title("Training Loss on Books-6 dataset for large and small networks")
axs.set_xlim(0, 49)
#axs.set_ylim(0, 13)
# Setup custom legend
# Guide: https://stackoverflow.com/questions/39500265/how-to-manually-create-a-legend
line = Line2D([0], [0], label='Default', linestyle='solid', color='k')
line_2 = Line2D([0], [0], label='Alternative', linestyle='dotted', color='k')
books_2_patch = mpatches.Patch(color='C0', label = '8 Layers')
books_3_patch = mpatches.Patch(color='C1', label = '2 Layers')
handles, labels = fig.gca().get_legend_handles_labels()
#handles.extend([books_2_patch])
plt.legend(handles=[line, line_2, books_2_patch, books_3_patch])
fig.savefig("../../data/results/july_7_books_6_breakthrough_sizes.png",bbox_inches='tight')
plt.show()
```

```python
df_reviews = pd.read_csv('../../data/results/july_7_results/reviews_loss_breakthrough.csv')
df_reviews = df_reviews.drop('Step', axis=1)
df_reviews = df_reviews.iloc[:, [int(x) for x in range(0, 12, 3)]]
```

```python
df_reviews.head()
```

```python
fig, axs = plt.subplots(figsize=(5,4))
df_reviews.plot.line(ax=axs, style={'edited_3_reviews_1024_8 - train_loss':':C0',
                                'edited_3_reviews_128_2 - train_loss':':C1',
                                'normal_reviews_1024_8 - train_loss':'C0',
                                'normal_reviews_128_2 - train_loss':'C1',})
axs.set_ylabel("Batch Loss")
axs.set_xlabel("Epochs")
axs.set_title("Training Loss on Reviews dataset for large and small networks")
axs.set_xlim(0, 49)
#axs.set_ylim(0, 13)
# Setup custom legend
# Guide: https://stackoverflow.com/questions/39500265/how-to-manually-create-a-legend
line = Line2D([0], [0], label='Default', linestyle='solid', color='k')
line_2 = Line2D([0], [0], label='Alternative', linestyle='dotted', color='k')
books_2_patch = mpatches.Patch(color='C0', label = 'Large networks (1024 hidden size; 8 layers)')
books_3_patch = mpatches.Patch(color='C1', label = 'Small networks (128 hidden size; 2 layers)')
handles, labels = fig.gca().get_legend_handles_labels()
#handles.extend([books_2_patch])
plt.legend(handles=[line, line_2, books_2_patch, books_3_patch])
fig.savefig("../../data/results/july_7_reviews_breakthrough_sizes.png",bbox_inches='tight')
plt.show()
```
