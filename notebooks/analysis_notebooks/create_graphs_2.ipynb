```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import matplotlib.patches as mpatches
```

```python
df_books_2 = pd.read_csv('../../data/results/july_7_results/books_2_loss.csv')
```

```python
df_books_2.head()
```

```python
df_books_2.shape
```

```python
# Drop pointless step column
df_books_2 = df_books_2.drop('Step', axis=1)
```

```python
df_books_2
```

```python
# We want every third column
columns = [x for x in range(0, 96, 3)]
print(columns)
```

```python
df_books_2 = df_books_2.iloc[:, [int(x) for x in range(0, 96, 3)]]
```

```python
df_books_2.head()
```

```python
# For this chart, we want to average the loss values for all networks to compare edited vs default
# We'll start by splitting into two dataframes to work with
default_df_books_2 = df_books_2.iloc[:, 16:]
alternative_df_books_2 = df_books_2.iloc[:, :16]
```

```python
# Average columns in both df
default_df_books_2['mean'] = default_df_books_2.mean(axis=1)
alternative_df_books_2['mean'] = alternative_df_books_2.mean(axis=1)
```

```python
default_df_books_2.head()
```

```python
# Time to plot
```

```python
fig, axs = plt.subplots(figsize=(5,3))
default_df_books_2['mean'].plot.line(ax=axs, style={'Mean':'C0'})
alternative_df_books_2['mean'].plot.line(ax=axs, style={'Mean':'C1'})
axs.set_ylabel("Batch Loss")
axs.set_xlabel("Epochs")
axs.set_title("Average training batch loss on 'Books-2' dataset")
axs.set_xlim(0, 49)
axs.set_ylim(0, 13)
plt.legend(['Default', 'Alternative'])
plt.show()
```
