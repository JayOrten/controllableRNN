```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.ticker import ScalarFormatter
import matplotlib.patches as mpatches
```

```python
df_books_3 = pd.read_csv('../../data/results/july_7_results/books_3_bert_hidden.csv')
```

```python
df_books_3.head()
```

```python
df_books_3 = df_books_3.drop('Step', axis=1)
df_books_3 = df_books_3.iloc[:, [int(x) for x in range(0, 24, 3)]]
```

```python
df_books_3.head()
```

```python
fig, axs = plt.subplots(figsize=(6,7))
df_books_3.plot.line(ax=axs, style={'edited_3_books_3_1024_6 - system_level_f1_BERT':':C0',
                                'edited_3_books_3_512_6 - system_level_f1_BERT':':C1',
                                'edited_3_books_3_256_6 - system_level_f1_BERT':':C2',
                                'edited_3_books_3_128_6 - system_level_f1_BERT':':C3',
                                'normal_books_3_1024_6 - system_level_f1_BERT':'C0',
                                'normal_books_3_512_6 - system_level_f1_BERT':'C1',
                                'normal_books_3_256_6 - system_level_f1_BERT':'C2',
                                'normal_books_3_128_6 - system_level_f1_BERT':'C3'})
axs.set_ylabel("BERTScore")
axs.set_xlabel("Epochs")
axs.set_title("BERTScore for various hidden sizes")
axs.set_xlim(0, 49)
#axs.set_ylim(0, 13)
# Setup custom legend
# Guide: https://stackoverflow.com/questions/39500265/how-to-manually-create-a-legend
line = Line2D([0], [0], label='Default', linestyle='solid', color='k')
line_2 = Line2D([0], [0], label='Alternative', linestyle='dotted', color='k')
patch1 = mpatches.Patch(color='C0', label = '1024 hidden size')
patch2 = mpatches.Patch(color='C1', label = '512 hidden size')
patch3 = mpatches.Patch(color='C2', label = '256 hidden size')
patch4 = mpatches.Patch(color='C3', label = '128 hidden size')
handles, labels = fig.gca().get_legend_handles_labels()
#handles.extend([books_2_patch])
plt.legend(handles=[line, line_2, patch1, patch2, patch3, patch4])
fig.savefig("../../data/results/july_7_books_3_hidden_size.png",bbox_inches='tight')
plt.show()
```

```python
df_books_3_ppl = pd.read_csv('../../data/results/july_7_results/books_3_ppl_hidden.csv')
df_books_3_ppl = df_books_3_ppl.drop('Step', axis=1)
df_books_3_ppl = df_books_3_ppl.iloc[:, [int(x) for x in range(0, 24, 3)]]
```

```python
df_books_3_ppl.head()
```
